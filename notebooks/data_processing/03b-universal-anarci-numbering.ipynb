# CDRH3 length inspection

```python
chen_deduplicated = pd.read_csv(path.join(DATA_DIR, "chen/deduplicated/chen_data.csv"), index_col=0)
len(chen_deduplicated)
```

```python
# Region of CDRH3 (positions 105 - 117):
heavy_df_chen.merge(chen_deduplicated["Antibody_ID"], left_on="Ab_ID", right_on="Antibody_ID")
cdrh3 = heavy_df_chen.merge(chen_deduplicated["Antibody_ID"], left_on="Ab_ID", right_on="Antibody_ID").iloc[:,131:196]
len(cdrh3)
```

```python
def is_aa(letter):
    if letter == "-":
        return 0
    else:
        return 1

def cdr_length(row):
    length = sum(is_aa(letter) for letter in row)
    return length
```

```python
cdrh3_length = cdrh3.apply(cdr_length, axis=1)
```

```python
len(cdrh3_length)
```
