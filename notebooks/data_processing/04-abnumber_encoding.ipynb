```python
from abnumber import Chain
import pandas as pd
import numpy as np
```

```python
chen_data = pd.read_csv("data/chen_data.csv", sep=";")
chen_data
```

```python
def split_ab(row):
    ab_seq = row["Antibody"]
    h, l = ab_seq.split(",")
    h = "".join(filter(str.isalnum, h))
    l = "".join(filter(str.isalnum, l))
    return h, l
```

```python
chains_df = chen_data.apply(split_ab, axis=1, result_type="expand")
chen_data["heavy"] = chains_df[0]
chen_data["light"] = chains_df[1]
chen_data.drop("Antibody", axis=1)
```

```python
chain = Chain(chen_data.iloc[1]["heavy"], scheme='imgt')
chain
```

```python
def chainify(seq):
    try:
        ch = Chain(seq, scheme="imgt")
        return ch
    except NotImplementedError as e:
        return None
    
```

```python
heavy = [chainify(h_seq) for h_seq in chen_data["heavy"]]
light = [chainify(l_seq) for l_seq in chen_data["light"]]
len(heavy)
```

```python
heavy_mask = [ch is not None for ch in heavy]
light_mask = [ch is not None for ch in light]
heavy_mask[:10]
```

```python
heavy_ids = chen_data["Antibody_ID"][heavy_mask]
light_ids = chen_data["Antibody_ID"][light_mask]
```

```python
heavy_clean = [ch for ch in heavy if ch is not None]
light_clean = [ch for ch in light if ch is not None]
```

```python
heavy_df = Chain.to_dataframe(heavy_clean)
heavy_df.insert(0, "Ab_ID", heavy_ids)
heavy_df
```

```python
light_df = Chain.to_dataframe(light_clean)
light_df.insert(0, "Ab_ID", light_ids)
light_df
```

```python
heavy_df.to_csv("data/chen_heavy_chain_numbering.csv", index=False)
light_df.to_csv("data/chen_light_chain_numbering.csv", index=False)
```

```python
AAs = "ACDEFGHIKLMNOPQRSTUVWY-"
aa_to_int = dict((c, i) for i, c in enumerate(AAs))
```

```python
def one_hot_encode_protein(seq):
    int_seq = [aa_to_int[aa] for aa in seq]
    onehot_encoded = []
    for val in int_seq:
        vect = [0 for i in range(len(AAs) - 1)]
        if val != 22:
            vect[val] = 1
        onehot_encoded += vect
    return np.asarray(onehot_encoded)
    
```

```python
def encode_proteins(data):
    res = data.apply(one_hot_encode_protein, axis=1, result_type="expand")
    return res
```

```python
def get_col_names(n):
    names = []
    for i in range(n):
        letters = [f"{ch}_{i + 1}" for ch in "ACDEFGHIKLMNOPQRSTUVWY"]
        names += letters
    return names
```

```python
encode_proteins(heavy_df.iloc[2:5,3:])
```

```python
heavy_one_hot = encode_proteins(heavy_df.iloc[:,3:])
heavy_one_hot.insert(0, "Ab_ID", heavy_df["Ab_ID"])
heavy_one_hot.columns = ["Ab_ID"] + get_col_names(len(heavy_df.columns) - 3)
heavy_one_hot.reset_index().to_feather("data/chen_heavy_one_hot.ftr")
```

```python
light_one_hot = encode_proteins(light_df.iloc[:,3:])
light_one_hot.insert(0, "Ab_ID", light_df["Ab_ID"])
light_one_hot.columns = ["Ab_ID"] + get_col_names(len(light_df.columns) - 3)
light_one_hot.reset_index().to_feather("data/chen_light_one_hot.ftr")
```
